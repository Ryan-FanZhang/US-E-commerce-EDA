**EDA -- Data Visualization **


```python
plt.figure(figsize = (20,10),dpi = 100)
# Check the relationship between the sale and the state
ax = sns.barplot(x = 'Sales', y = 'State', data= df, palette="Blues_d")
plt.xlabel('Sales',fontsize = '40')
plt.ylabel('State',fontsize = '40')
plt.tight_layout()
```

```python
plt.figure(figsize = (20,10),dpi = 100)
# Check the relationship between the sale and the region 
ax = sns.barplot(x = 'Sales', y = 'Region', data= df)
plt.xlabel('Sales',fontsize = '40')
plt.ylabel('State',fontsize = '40')
plt.tight_layout()
```

```python
# Check the sales andd sub_category
plt.figure(figsize = (20,10), dpi = 100)
sns.barplot(x=df['Sub-Category'], y=df['Sales'],palette = "Blues")
plt.xlabel('Sub-Category')
plt.ylabel('Sales')
```

```python
# Check the sales andd category
plt.figure(figsize = (20,10), dpi = 100)
sns.barplot(x=df['Category'], y=df['Sales'],palette = "Blues")
plt.xlabel('Category')
plt.ylabel('Sales')
```

```python
# Pair plot for a region wise see whether a bit overwhelming ?
plt.figure(figsize=(20,10))
sns.pairplot(df,hue="Region")
```

```python
# Checking which region could make high profit?
plt.figure(figsize = (20,10), dpi = 100)
sns.barplot(x=df['Region'], y=df['Profit'],palette = "Blues")
plt.xlabel('Region')
plt.ylabel('Profit')
```

```python
# Checking whether the discount could make high profits
profit_discount = df.groupby('Discount')['Profit'].sum()
plt.figure(figsize=(20,10))
sns.barplot(x=profit_discount.index,y=profit_discount.values)
plt.xlabel('Discount')
plt.ylabel('Profit')
```

```python
# Checking the replationship between state and loses
losses = df.loc[df['Profit']<=0]
losses.head()
```

```python
losses['Profit'] = losses['Profit'].abs()
```

```python
state_losses= losses.groupby('State')['Profit'].sum()
```

```python
plt.figure(figsize=(50,20))
state_losses.plot.pie(autopct="%.1f%%")
```

```python
# Checking the replationship between sub_category and loses
sub_category_losses = losses.groupby('Sub-Category')['Profit'].sum()
plt.figure(figsize=(50,20))
sub_category_losses.plot.pie(autopct="%.1f%%")
```

```python
# Checking the replationship between category and loses
sub_category_losses = losses.groupby('Category')['Profit'].sum()
plt.figure(figsize=(50,20))
sub_category_losses.plot.pie(autopct="%.1f%%")
```

```python
# Checking the replationship between segments and loses
sub_category_losses = losses.groupby('Segment')['Profit'].sum()
plt.figure(figsize=(50,20))
sub_category_losses.plot.pie(autopct="%.1f%%")
```
